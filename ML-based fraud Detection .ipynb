```python
import pandas as pd          
import numpy as np
```

```python
from sklearn.preprocessing import StandardScaler    
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
#Converting CSV file into dataframes
card_data=pd.read_csv('creditcard.csv') 
```

```python
#dataset_information
card_data.info()
```

```python
 card_data.head()
```

```python
#checking the number of missing values in each column
card_data.isnull().sum()
```

```python
card_data['Class'].value_counts()
```

```python
authorized=card_data[card_data.Class == 0]
print(authorized)
```

```python
fraud=card_data[card_data.Class == 1]
print(fraud)
```

```python
print(authorized.shape)
print(fraud.shape)
```

```python
#Statistical measures of the data
authorized.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compare the value for both transactions
card_data.groupby('Class').mean()
```

```python
#Number of fraudulent Transactions --->492
authorized_sample=authorized.sample(n=492)
print(authorized_sample.shape)
print(authorized_sample)
```

```python
#Concatenating the two dataframes
new_dataset=pd.concat([authorized_sample,fraud],axis=0)
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
print(X.shape)
print(Y.shape)
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
model=LogisticRegression()
```

```python
#training the logistic Regression Model with Training data
model.fit(X_train,Y_train)
```

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)

print(training_data_accuracy)
```

```python
#accuracy on Testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)

print(testing_data_accuracy)
```
